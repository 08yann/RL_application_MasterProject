## Try DQN algorithm with Binomial simulated data with first stock higher probability of going up

```python
simplebin_w=SimpleBinomial(1.02,0.98,p1=0.52,p2=0.5)
Nsample=100000
ret_w=simplebin_w.simulate(Nsample+1)
plt.plot(np.cumprod(ret_w[0,:]))
plt.plot(np.cumprod(ret_w[1,:]))
plt.legend(['S1','S2'])
plt.show()
```

```python
true_qval_w=simplebin_w.true_qval()
```

```python
window=1
decay_rate=0.1
gamma=0.98
pol_w=Policy(window,gamma)
a0=pol_w.train(pd.DataFrame(ret_w,index=['S1','S2']),true_qval_w,decay_rate=0.1,timestep_error=1000)

```

```python
ret_w=pd.DataFrame(ret_w, index=['S1','S2'])

Nsim=10000
test_ret_w=simplebin_w.simulate(Nsim)
```

```python
actor=DQN2(ret_w,batch_size=32,nb_epochs=100,lr=0.1)
actor.train()
res_dqn=analysis_result(actor,np.array(test_ret_w))
plot_result(res_dqn)

err=actor.compare_DQN2_discrete(true_qval_w)
plot_error_per_state(err)
```

```python
actor=DQN2(ret_w,batch_size=256,nb_epochs=100,lr=0.01)
actor.train()
res_dqn=analysis_result(actor,np.array(test_ret_w))
plot_result(res_dqn)

err=actor.compare_DQN2_discrete(true_qval_w)
plot_error_per_state(err)
```

```python
actor=DQN2(ret_w,batch_size=64,nb_epochs=100,lr=0.001)
actor.train()
res_dqn=analysis_result(actor,np.array(test_ret_w))
plot_result(res_dqn)

err=actor.compare_DQN2_discrete(true_qval_w)
plot_error_per_state(err)
```

# Binomial correlated processes ($\rho = 0.05$) delayed by two time-steps

```python
np.random.seed(72349)
rho=0.05
simplebin_corrdel05=SimpleBinomial(1.02,0.98,0.5,corr=rho,delay=2)
Nsample=100000
ret_corrdel05=simplebin_corrdel05.simulate(Nsample+1)
plt.plot(np.cumprod(ret_corrdel05[0,:]))
plt.plot(np.cumprod(ret_corrdel05[1,:]))
plt.legend(['S1','S2'])
plt.show()
```

### Convergence doesn't happen.

However we see for some choices of parameters a beginning of convergence. A larger training data would lead to convergence, since the relation of a correlation $\rho=0.05$ is too small for only $100'000$ time-steps.

# Delayed Markov process

We simulate $r_{1,t}$ using a Markov chain with transition matrix $p_{rand}$ and set the second return as:

$$r_{2,t} = r_{1,t+2}, \quad \forall t \geq 0 $$

```python
np.random.seed(1938)
p_rand = np.random.uniform(size=(7,7))
for i in range(p_rand.shape[0]):
    p_rand[i]=np.exp(p_rand[i])/np.sum(np.exp(p_rand[i]))
np.sum(p_rand)

p_rand
```

```python
w2=2
delta=0.01
gamma=0.98
delay=2
MC_rand_del2=MarkovChain_delayed(delta, p_rand,delay)
train_del2=pd.DataFrame(MC_rand_del2.simulate(100000+1),index=['S1','S2'])
pol_delay2=Policy(w2,gamma)
```

```python
true_qval_delay2=MC_rand_del2.true_qval(delta=1e-3)
```

```python
Ntest=1000
test_delay2=pd.DataFrame(pd.DataFrame(MC_rand_del2.simulate(Ntest),index=['S1','S2']),index=['S1','S2'])

```

```python
PROCESSES = 6
with multiprocessing.Pool(PROCESSES) as pool:
    params=list()
    for lr in [0.01,0.001,0.0001]:
        for batch_size in [64,256]:
            params.extend([create_param_DQN(pd.DataFrame(train_del2),test_delay2,window=2,lr=lr,batch_size=batch_size,nb_epochs=100,true_qval=true_qval_delay2)])
            
    jobs_delay2 = pool.map(multi_train, params)
```
