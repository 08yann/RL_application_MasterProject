```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import pandas as pd

import discrete_sim_methods
from discrete_sim_methods import simulation_geom, Policy, discretize

np.random.seed(19)
# Create simulated stocks as arithmetic Brownian motions : dS = \mu dt + \sigma_1 dW_1t 
dt = 1/252
# 252 as there are approximately 252 trading days in a year
N=100000
mu = np.array([[0.02,0.01]]).transpose()
sig = np.array([[0.2,0.1]]).transpose()
S0=10

stocks=simulation_geom(N, S0, mu, sig)
stocks_discrete=discretize(stocks)
```

# Equal weight benchmark

```python
eq_weight=0.5*np.ones((2,1))
equal_w_portfolio=np.matmul(eq_weight.transpose(),stocks)
plt.plot(equal_w_portfolio.transpose())
plt.show()


ret_equalw_port= 10*np.prod(equal_w_portfolio.iloc[0,:])
print("Final value of equal weight portfolio: ", ret_equalw_port)
```

```python
computer=Policy()
window=2

computer.train(stocks_discrete)
best_train=computer.exploit(stocks_discrete)
port_train=np.sum(best_train* stocks.iloc[:,window+1:])

plt.plot(port_train)
plt.show()

totalret_best_port= 10*np.prod(port_train)
print(totalret_best_port)
```

```python
plt.plot(10*np.cumprod(port_train))
plt.title("Evolution of trained portfolio")
plt.show()

plt.plot(10*(np.cumprod(np.array(equal_w_portfolio))),color="red")
plt.title("Evolution of equal weight portfolio")
plt.show()
```

# Correlated Brownian motions for simulation

We suppose Corr( W1_t, W2_t) = rho.

This can be implemented simply with: dW2_t = rho dW1_t + \sqrt(1-rho^2) dB_t

where B_t independent from W1_t.

```python
def simulate_correlated_BM(N, S0,mu, sigma, rho=0):
    dt=1/252
    W1 = np.random.normal(size=(1, N))
    W2 = rho *W1 + np.sqrt(1-rho**2)* np.random.normal(size=(1, N))
    W=np.vstack((W1,W2))

    S = S0 * np.exp(np.cumsum((mu - 0.5 * sigma ** 2) * dt + sigma * W * np.sqrt(dt), axis=1))

    # S=10+np.cumsum(mu*dt_sim + np.sqrt(dt)*sig*W,axis=1)
    R = S[:, 1:] / S[:, :-1]
    plt.plot(S.transpose())
    plt.legend(["S" + str(i) for i in range(1, mu.shape[0]+1)])
    plt.show()
    stocks = pd.DataFrame(R, index=["S" + str(i) for i in range(1, mu.shape[0]+1)])

    return stocks

mu = np.array([[0.02,0.01]]).transpose()
sig = np.array([[0.2,0.1]]).transpose()
S0=10
rho=0.8
stocks_corr=simulate_correlated_BM(N, S0, mu, sig,rho)
```

```python
# Compare with equal weight
train_time_corr=int(np.ceil(0.8*stocks_corr.shape[1]))
train_data_corr=discretize(stocks_corr.iloc[:,:train_time_corr])
test_data_corr=discretize(stocks_corr.iloc[:,train_time_corr:])

equal_w_portfolio_corr=np.matmul(eq_weight.transpose(),stocks_corr.iloc[:,train_time_del+window_3+1:])
plt.plot(10*np.cumprod(np.array(equal_w_portfolio_corr.transpose())),color="red")
plt.show()


ret_equalw_port_corr= 10*np.prod(equal_w_portfolio_corr.iloc[0,:])
print("Final value of equal weight portfolio: ", ret_equalw_port_corr)
```
