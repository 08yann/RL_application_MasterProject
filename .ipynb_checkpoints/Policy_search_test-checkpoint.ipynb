```python
class DDPG_eachtimestep:
    def __init__(self, train_data, window=1, batch_size=32, gamma=0.98, actor_lr=0.001, critic_lr=0.001, tau=0.005, units_actor =256,units_critic=256,noise_param = 0.01):
        self.window = window
        self.actor_lr = actor_lr
        self.critic_lr = critic_lr
        self.nb_asset = train_data.shape[0]
        self.batch_size = batch_size

        self.train_data = train_data
        self.gamma = gamma

        self.tau = tau
        self.units_actor = units_actor
        self.units_critic = units_critic

        self.actor_model = self.build_actor_model()
        self.critic_model = self.build_critic_model()

        self.target_actor = self.build_actor_model()
        self.target_critic = self.build_critic_model()
        self.target_actor.set_weights(self.actor_model.get_weights())
        self.target_critic.set_weights(self.critic_model.get_weights())

        self.critic_optimizer = tf.keras.optimizers.Adam(self.critic_lr)
        self.actor_optimizer = tf.keras.optimizers.Adam(self.actor_lr)


        self.noise_param = noise_param

    def build_actor_model(self):
        inp = tf.keras.layers.Input(shape=(self.nb_asset * self.window,), name="Prev_ret")
        lay = tf.keras.layers.Dense(units=self.units_actor, activation='relu', name='Dense_1')(inp)
        lay = tf.keras.layers.Dense(units=self.units_actor, activation='relu', name='Dense_2')(lay)
        weights = tf.keras.layers.Dense(units=2, activation='softmax', name='softmax')(lay)

        model = tf.keras.models.Model(inp, weights)
        return model

    def build_critic_model(self):
        inp_state = tf.keras.layers.Input(shape=(self.nb_asset * self.window,), name="Prev_ret")
        lay_state = tf.keras.layers.Dense(units=int(np.round(self.units_critic/4)), activation='relu', name='Dense_1')(inp_state)
        lay_state = tf.keras.layers.Dense(units=int(np.round(self.units_critic/2)), activation='relu', name='Dense_2')(lay_state)

        inp_weight = tf.keras.layers.Input(shape=(2), name="Action")
        lay_weight = tf.keras.layers.Dense(units=int(np.round(self.units_critic/2)), activation='relu')(inp_weight)

        concat = tf.keras.layers.Concatenate()([lay_state, lay_weight])

        q_val = tf.keras.layers.Dense(units=self.units_critic, activation='relu')(concat)
        q_val = tf.keras.layers.Dense(units=self.units_critic, activation='relu')(q_val)
        q_val = tf.keras.layers.Dense(units=1,activation='relu')(q_val)

        model = tf.keras.models.Model([inp_state, inp_weight], q_val)
        return model

    def reward(self, weight, new_ret):
        #         print(np.dot(weight.reshape(-1,1).transpose(), new_ret))
        return np.log(np.dot(weight.reshape(-1, 1).transpose(), new_ret))

    def policy(self, state):
        sampled_action = tf.squeeze(self.actor_model(state.reshape(1, -1)))
        noise = self.noise_param * np.random.randn()
        noise_vec = np.array([noise, -noise])
        sample_action = sampled_action.numpy() + noise_vec
        action = np.clip(sample_action, 0, 1)

        return action

    # @tf.function
    def update_target_actor(self):
        self.target_actor.set_weights(self.actor_model.get_weights())
        # for (a, b) in zip(self.target_actor.variables, self.actor_model.variables):
        #     a.assign(b * self.tau + a * (1 - self.tau))

    # @tf.function
    def update_target_critic(self):
        self.target_critic.set_weights(self.critic_model.get_weights())

        # for (a, b) in zip(self.target_critic.variables, self.critic_model.variables):
        #     a.assign(b * self.tau + a * (1 - self.tau))

    # @tf.function
    def train(self):
        nb_episode = int(np.floor(self.train_data.shape[1]) / self.batch_size)

        for e in range(nb_episode - 1):
            prev_states = np.array(self.train_data.iloc[:, e * self.batch_size:(e + 1) * self.batch_size]).transpose()
            next_states = np.array(
                self.train_data.iloc[:, e * self.batch_size + 1:(e + 1) * self.batch_size + 1]).transpose()

            action_batch = np.zeros((2, self.batch_size))
            reward_batch = np.zeros(self.batch_size)
            q_val_target = np.zeros((self.batch_size, 1))

            for i in range(self.batch_size):
                action_batch[:, i] = self.policy(prev_states[i, :])
                reward_batch[i] = self.reward(action_batch[:, i], next_states[i, :].transpose())

            #             print(next_states.shape)

                with tf.GradientTape() as tape:
                    target_actions = self.target_actor(next_states[i,:].reshape(1, -1), training=True).numpy()
                    q_val_target = reward_batch[i] + self.gamma *self.target_critic([next_states[i,:].reshape(1, -1), target_actions.reshape(1, -1)],
                                                                                  training=True)

                    critic_val = self.critic_model([prev_states[i,:].reshape(1, -1), action_batch[:,i].reshape(1, -1)], training=True)
                    critic_loss = tf.math.reduce_mean(tf.math.square(q_val_target - critic_val))

                    critic_grad = tape.gradient(critic_loss, self.critic_model.trainable_variables)
                    self.critic_optimizer.apply_gradients(zip(critic_grad, self.critic_model.trainable_variables))

                with tf.GradientTape() as tape:
                    actions = self.actor_model(next_states, training=True)
                    critic_val = self.critic_model([prev_states, actions], training=True)

                    actor_loss = -tf.math.reduce_mean(critic_val)

                    actor_grad = tape.gradient(actor_loss, self.actor_model.trainable_variables)
                    self.actor_optimizer.apply_gradients(zip(actor_grad, self.actor_model.trainable_variables))

            self.update_target_actor()
            self.update_target_critic()






```

```python
diag_plus=np.array([1,.1,.5,.8,.6])
diag_minus=np.array([.9,.5,.2, .4,1])
p=np.diag(diag_plus,1)+np.diag(diag_minus,-1)
delta=0.01

np.random.seed(1938)
p_rand = np.random.uniform(size=(7,7))
for i in range(p_rand.shape[0]):
    p_rand[i]=np.exp(p_rand[i])/np.sum(np.exp(p_rand[i]))
np.sum(p_rand)

delta=0.01
MC2d=MarkovChain_2D(delta,p,p_rand)

N=100000
training=pd.DataFrame(MC2d.simulate(N),index=['S1','S2'])

N_test=1000
testing=pd.DataFrame(MC2d.simulate(N_test),index=['S1','S2'])



```

```python
import gc
for bs in [64,256,1024]:
    for tau in [0.1,0.01,0.005]:
        for units_actor in [64,256]:
            for noise_param in [0.01,0.1,0.001]:
                pol_search_MC = DDPG(pd.DataFrame(training,index=['S1','S2']),batch_size=bs,tau=tau, units_actor=units_actor,noise_param=noise_param)
                pol_search_MC.train()

                w=pol_search_MC.actor_model(np.array(testing.iloc[:,:-1]).transpose()).numpy()
                plt.plot(np.cumprod(testing.iloc[0,:]))
                plt.plot(np.cumprod(testing.iloc[1,:]))
                plt.plot(np.cumprod(np.sum(w.transpose()*testing.iloc[:,1:],axis=0)))
                plt.legend(['S1','S2','Learned'])
                plt.show()

                plt.hist(w[:, 0], bins=np.arange(0, 12) / 10)
                plt.title('Histogram of weights in stock 1')
                plt.xlabel('Action')
                plt.show()
                gc.collect()
```

```python
for bs in [32,64,128,256]:
    pol_search_MC = DDPG(pd.DataFrame(np.log(training),index=['S1','S2']),batch_size=bs,units_actor =32,units_critic=32,tau=0.1)
    pol_search_MC.train()

    w=pol_search_MC.actor_model(np.array(np.log(testing.iloc[:,:-1])).transpose()).numpy()
    plt.plot(np.cumprod(testing.iloc[0,:]))
    plt.plot(np.cumprod(testing.iloc[1,:]))
    plt.plot(np.cumprod(np.sum(w.transpose()*testing.iloc[:,1:],axis=0)))
    plt.legend(['S1','S2','Learned'])
    plt.show()

    plt.hist(w[:, 0], bins=np.arange(0, 12) / 10)
    plt.title('Histogram of weights in stock 1')
    plt.xlabel('Action')
    plt.show()
```
