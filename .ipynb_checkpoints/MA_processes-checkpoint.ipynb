AR processes have not directly values close to zero and thus if we want to simulate returns of some stock, it isn't straightforward. One possibility is to divide the simulation by $100$ and add $1$, thus we get return between $1.04$ and $0.96$.

However, MA processes can be almost directly used to simulate returns by simpling adding $1$ to the values.

```python
ma1=MAProcess([.01],mean=0.01)
ma2=MAProcess([.01],mean=0.02)


ma_data=np.array([1+ma1.simulate(10000),1+ma2.simulate(10000)])
plt.plot(ma_data[0])
plt.plot(ma_data[1])
plt.show()
```

# 2 Stock's returns as MA processes

We use MA processes to simulate returns. We still use same QLearning algorithm for tabular case, thus we also need to implement a discretizing routine. We bound return at the second decimal, i.e. at $0.01$ with minimum and maximum: $0.96$ and $1.04$.

```python
Nsim=1000000
window=1
gamma=0.98
pol=Policy(window,gamma)
```

```python
ma2d=MAProcess2d([0.01])
```

```python
ma2d_data=pd.DataFrame(ma2d.simulate(Nsim),index=['S1','S2'])
```

```python
ma2d.discrete_true_qval()
```

```python
pol.train(discretize(ma2d_data))
```

```python
pol.q_val
```

```python
Ntest=10000
ma2d_test=ma2d.simulate(Ntest)


res=analysis_result(pol,np.array(ma2d_test))
plot_result(res)
```

```python
pol.q_val
```

# Return for stock 1 has small positive mean

```python
ma2dm=MAProcess2d([0.01],np.array([[0.01],[0.]]))
ma2dm_data=ma2dm.simulate(Nsim)
polm=Policy(window)
polm.train(discretize(ma2dm_data))
```

```python
Ntest=10000
ma2dm_test=ma2dm.simulate(Ntest)

resm=analysis_result(polm,np.array(ma2dm_test))
plot_result(resm)
```

```python
polm.q_val
```

Optimal policy puts always weight $1$ in stock $1$ since it has positive mean.

```python
w2=2
polm2=Policy(w2)
polm2.train(discretize(ma2dm_data))
```

```python
resm2=analysis_result(polm2,np.array(ma2dm_test))
plot_result(resm2)
```

```python
w2=2
pol_corr2=Policy(w2)
pol_corr2.train(discretize(ma_corrtrain))
```

```python
pol_corr2.q_val
```

```python
res_corr2=analysis_result(pol_corr2,np.array(ma_corrtest))
plot_result(res_corr2)
```

```python
plt.plot(np.cumprod(ma_corrtest.iloc[0,:]))
plt.plot(np.cumprod(ma_corrtest.iloc[1,:]))
plt.plot(np.cumprod(0.5*ma_corrtest.iloc[0,:]+0.5*ma_corrtest.iloc[1,:]),color='red')

plt.legend(['S1','S2','Eq. weight'])
plt.show()
```

Since we are looking at a MA(2) process, looking at window=2 leads to really good policy, close to optimality. Even with only window=1, we still pick up some knowledge about the process.

# Delay second return by one time-step

Since both MA are simulated with correlated gaussian white noise, delaying the second return creates a time dependence.

We use exactly same data as ma_corrtrain and ma_corrtest, only delayed for second return.

$$ r_{2,t} = r_{1,t+1}$$

Hence, optimal policy always chooses weight $w_1=1$ if log-return of second stock is negative and $w_1=0$ otherwise.

```python
ma_corrtrain_del=np.array(ma_corrtrain.iloc[:,:-1])
ma_corrtrain_del[1,:]= np.array(ma_corrtrain.iloc[1,1:])
ma_corrtest_del=np.array(ma_corrtest.iloc[:,:-1])
ma_corrtest_del[1,:]=np.array(ma_corrtest.iloc[1,1:])
```

```python
pol_corrdel=Policy()
pol_corrdel.train(discretize(pd.DataFrame(ma_corrtrain_del, index=['S1','S2'])))
```

```python
res_corrdel=analysis_result(pol_corrdel,ma_corrtest_del)
plot_result(res_corrdel)
```

```python
for k in pol_corrdel.q_val.keys():
    print(k, np.argmax(pol_corrdel.q_val[k]))
```

Almost optimal policy.

```python
res_corrdel2=analysis_result(pol_corrdel2,ma_corrtest_del)
plot_result(res_corrdel2)
```
