```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(129)


# Create simulated stock as arithmetic Brownian motions : dS = \mu dt + \sigma dWt 
dt = 1/252
# 252 as there are approximately 252 trading days in a year
# Initial price S0, mean \mu, standard deviation \sigma
S0=10
mu = 0.1
sig = 0.2
N=252*10

W= np.random.randn( N)

S=S0+np.cumsum(mu*dt + np.sqrt(dt)*sig*W)
plt.plot(S)
plt.title("Stock price")
plt.show()

R=np.diff(S)/S[:-1]
plt.plot(R)
plt.title("Stock price")
plt.show()


# Risk free interest rate:
# We want to know daily interest rate corresponding to 0.01 yearly
daily_rf=(1.01**(1/252)-1)*252
rf=10*np.ones(N)
rf[1:]=10*np.cumprod((1+daily_rf/252)*np.ones(N-1))
```
