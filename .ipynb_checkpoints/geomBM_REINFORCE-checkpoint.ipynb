```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import pandas as pd

import discrete_sim_methods
from discrete_sim_methods import simulation_geom, Policy, discretize

np.random.seed(19)
# Create simulated stocks as arithmetic Brownian motions : dS = \mu dt + \sigma_1 dW_1t 
dt = 1/252
# 252 as there are approximately 252 trading days in a year
N=100000
mu = np.array([[0.02,0.01]]).transpose()
sig = np.array([[0.2,0.1]]).transpose()
S0=10

stocks=simulation_geom(N, S0, mu, sig)
stocks_discrete=discretize(stocks)
```

```python
eq_weight=0.5*np.ones((2,1))
equal_w_portfolio=np.matmul(eq_weight.transpose(),stocks)
plt.plot(10*np.cumprod(equal_w_portfolio.transpose()))
plt.show()


ret_equalw_port= 10*np.prod(equal_w_portfolio.iloc[0,:])
print("Final value of equal weight portfolio: ", ret_equalw_port)
```

# Equal weight portfolio benchmark

```python
train_time=int(np.ceil(0.8*stocks_discrete.shape[1]))
train_data=stocks_discrete.iloc[:,:train_time]
test_data=stocks_discrete.iloc[:,train_time:]

eq_weight=0.5*np.ones((2,1))
equal_w_portfolio=np.matmul(eq_weight.transpose(),test_data)
plt.plot(10*np.cumprod(equal_w_portfolio.transpose()),color='red')
plt.show()

ret_equalw_port= 10*np.prod(equal_w_portfolio.iloc[0,:])
print("Final value of equal weight portfolio: ", ret_equalw_port)
```

```python
import keras
import tensorflow as tf
class Agent_reinforce():
    def __init__(self,train_data,nb_asset=2,nb_action=11):
        self.nb_asset=nb_asset
        self.nb_action=nb_action
        self.model=self.build_model()
        self.train_data=train_data
        self.gamma=0.98
        
    def build_model(self):
        inp=tf.keras.layers.Input(shape=(self.nb_asset,),name="Prev_ret")
        lay1=tf.keras.layers.Dense(units=12,activation='tanh',name='Dense_1')(inp)
        lay2=tf.keras.layers.Dense(units=32,activation= 'tanh',name='Dense_2')(lay1)
        q_val=tf.keras.layers.Dense(units=self.nb_action,activation='linear',name='q_val')(lay2)


        optimizer=tf.keras.optimizers.Adam()

        model=tf.keras.models.Model(inp,q_val)
        model.compile(loss='mse',optimizer=optimizer)
        model.summary()
        return model
                    
        
    def chooseW(self,state):
        q_val=self.model.predict(state)[0]
        print(q_val)
        w=np.argmax(q_val)/(self.nb_action-1)
        return w
    
    def train(self):
        batch_size=1000
        
        nb_episode=int(np.floor(self.train_data.shape[1])/batch_size)
        
        for e in range(nb_episode-1):
            update_input=np.zeros((batch_size,self.nb_asset))

            update_nextstate=np.zeros((batch_size,self.nb_asset))

            list_action=[]

            list_rew=[]

            for i in range(batch_size):
                update_input[i]=self.train_data.iloc[:,e*batch_size+i]
                update_nextstate[i]=self.train_data.iloc[:,e*batch_size+i]

                list_action.append(np.random.randint(11))
                list_rew.append((np.dot(np.array([[list_action[i]/(self.nb_asset-1)],[1-list_action[i]/(self.nb_asset-1)]]).transpose(),update_nextstate[i]))[0])

            q_val=self.model.predict(update_input)
            q_val_next=self.model.predict(update_nextstate)
            

            for i in range(batch_size):
                q_val[i][list_action[i]]=list_rew[i]+self.gamma * np.argmax(q_val_next[i])
            

            
            self.model.fit(update_input, q_val, batch_size=batch_size, epochs=1, verbose=0)
        
            
            
            
    def exploit(self,test_data,init_inv=100):
        w=np.zeros((2,test_data.shape[1]-1))
        w[0,:]=self.model.predict(test_data.iloc[:,:-1])/(self.nb_asset-1)
        w[1,:]=1-w[0,:]
        
        ret=np.sum(w*test_data.iloc[:,1:])
        plt.plot(np.cumprod(ret))
        plt.show()
        
        return ret
        
        
        
```

```python
train_stocks=stocks.iloc[:,:train_time]
test_stocks=stocks.iloc[:,train_time:]
pol=Agent_reinforce(train_stocks)
```

```python
pol.train()
```

```python
pol.model.get_weights()
```

```python
w=np.zeros((2,test_stocks.shape[1]-1))

w[0,:]=np.argmax(pol.model.predict(np.array(test_stocks.iloc[:,:-1]).transpose()),axis=1)/10
w[1,:]=1-w[0,:]

ret=np.sum(w*test_stocks.iloc[:,1:])
plt.plot(np.cumprod(ret))
plt.show()
        
    
total_ret_DQN= 10*np.prod(ret)
print("Final value of equal weight portfolio: ", total_ret_DQN)
```
