```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import pandas as pd
import yfinance as yf
import json
from yahoofinancials import YahooFinancials

#us_equities=pd.read_csv("data/us_equities.csv.gz")

with open("data/company_tickers.json") as f:
    tickers= json.load(f)
```

```python
tickers_list=[]
for i in range(120):
    tickers_list.append(tickers[str(i)]['ticker'])
tickers_list.insert(0,'^GSPC')

stocks=pd.DataFrame(yf.Ticker(tickers_list[0]).history(period='2y',interval='1h')['Close']).rename(columns={'Close' : tickers_list[0]},inplace=False)
for i in range(1,len(tickers_list)):
    temp=pd.DataFrame(yf.Ticker(tickers_list[i]).history(period='2y',interval='1h')['Close']).rename(columns={'Close' : tickers_list[i]},inplace=False)
    stocks=stocks.merge(temp,how = 'outer', right_index=True, left_index=True)

stocks=stocks.drop(stocks.tail(1).index)
stocks=stocks.drop(stocks[stocks['AAPL'].isna()].index)    
stocks=stocks.drop(np.sum(stocks.isna())[np.sum(stocks.isna())>0].index,axis=1)
```

```python
PROCESSES = 6
with multiprocessing.Pool(PROCESSES) as pool:
    params=list()
    for decay_rate in [100,10,1]:
        for tau in [1,0.8]:
            params.extend([create_param_QLearning(pd.DataFrame(discretize(ret_2[:,:3000]),index=['S1','S2']) ,np.array(discretize(ret_2[:,3000:])),  window=1, decay_rate=decay_rate,tau=tau,adaptive_lr=True)])
            
    res_QLearning = pool.map(multi_train_QLearning, params)
```
